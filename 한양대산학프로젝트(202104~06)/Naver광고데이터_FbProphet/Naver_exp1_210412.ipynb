```python
import time
import random
import requests
import json
import signaturehelper
import pandas as pd
from pandas import DataFrame

def get_header(method, uri, api_key, secret_key, customer_id):
    timestamp = str(round(time.time() * 1000))
    signature = signaturehelper.Signature.generate(timestamp, method, uri, SECRET_KEY)
    return {'Content-Type': 'application/json; charset=UTF-8', 'X-Timestamp': timestamp,
          'X-API-KEY': API_KEY, 'X-Customer': str(CUSTOMER_ID), 'X-Signature': signature}

BASE_URL = 'https://api.naver.com'
API_KEY = '01000000004515f5651224e4b84e5ea0eef7e4d6ed9f1f34e45950585526be315034d19c57'
SECRET_KEY = 'AQAAAABFFfVlEiTkuE5eoO735Nbt6Agzn/RajmeUL626d68fVA=='
CUSTOMER_ID = '2060998'    #smartmind

uri = '/ncc/adgroups'
method = 'GET'
r = requests.get(BASE_URL + uri, headers=get_header(method, uri, API_KEY, SECRET_KEY, CUSTOMER_ID))
rj = json.dumps(r.json(), indent="\t", ensure_ascii= False)
print("response status_code = {}".format(r.status_code))
print("response body = {}".format(rj))

target_adgroup = r.json()
```

```python
## 캠페인 전체

nCID = []
for elem in target_adgroup:
    if elem['status'] == 'ELIGIBLE':
        nCID.append(elem['nccCampaignId'])
nCID
```

```python
import datetime
from datetime import date

since = date(2021, 3, 31)
until = date(2021, 4, 11)
delta = until - since

daylist=[]
timerange=[]
for i in range(delta.days+1):
    day_next = since + datetime.timedelta(days=i)
    days = day_next.strftime("%Y-%m-%d")
    daylist.append(days)
    during = '{'+'"since":"'+days+'", "until":"'+days+'"}'
    timerange.append(during)
```

```python
uri = '/stats'
method = 'GET'
stat_ids = nCID

rlist = []

for during in timerange:
    print(during)
    r = requests.get(BASE_URL + uri,
                     params={'ids': nCID,
                             'fields': '["impCnt", "clkCnt", "salesAmt", "ctr", "cpc", "avgRnk", "ccnt", "pcNxAvgRnk", "mblNxAvgRnk", "crto", "convAmt", "ror", "cpConv"]',
                             'timeRange': during,                         
                             'breakdowns': 'hh24'},
                     headers=get_header(method, uri, API_KEY, SECRET_KEY, CUSTOMER_ID))
    rlist.append(r)
    rj3 = json.dumps(r.json(), indent="\t", ensure_ascii= False)
    print("response status_code = {}".format(r.status_code))
    print("response body = {}".format(rj3))
rlist2 = list(map(lambda i: i.json(), rlist))
```

```python
fields=rlist2[0]['data'][0]['breakdowns'][0].keys()
df=DataFrame(None,columns=fields)
day_list = []
df_day=DataFrame()
id_day=DataFrame()
cid_list=[]

for n in range(len(rlist2)):
    day = rlist2[n]
    date = daylist[n]

    for ids in range(len(rlist2[n]['data'])):
        cid = rlist2[6]['data'][ids]['id']

        for idx in range(0, len(day['data'][ids]['breakdowns'])):
            df=df.append(DataFrame(day['data'][ids]['breakdowns'][idx],index=[idx]))
            day_list.append(date)
            cid_list.append(cid)

df_data = df.reset_index().drop('index', axis=1)
df_day=df_day.append(DataFrame(day_list, columns=['날짜']))
id_day=id_day.append(DataFrame(cid_list, columns=['ID']))
df_all = pd.concat([df_day, id_day, df_data], axis=1)

df_all.rename({'name':'시간','ctr':'클릭률','clkCnt':'클릭수','cpc':'평균클릭비용','ccnt':'전환수','avgRnk':'평균노출순위','impCnt':'노출수','salesAmt':'총비용'}, axis=1, inplace=True)
df_all.set_index('날짜',inplace=True)

df_all=df_all[['ID', '시간', '노출수', '클릭수', '클릭률', '평균클릭비용', '총비용','전환수', '평균노출순위']]
df_all
```
