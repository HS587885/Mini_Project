```python
df_imp = df[['date','impressions']]
df_imp
```

```python
ar_imp = df_imp.rename(columns={'date':'ds', 'impressions':'y'})
ar_imp.tail()
```

```python
df_clicks = df[['date', 'clicks']]
df_clicks.tail()
```

```python
ar_clicks = df_clicks.rename(columns={'date':'ds', 'clicks':'y'})
ar_clicks.tail()
```

```python
from fbprophet import Prophet

prophet = Prophet(seasonality_mode='multiplicative', 
                  yearly_seasonality=False,
                  weekly_seasonality=True, daily_seasonality=False,
                  changepoint_prior_scale=0.5)
prophet.fit(ar_clicks)
```

```python
future_data = prophet.make_future_dataframe(periods=3, freq='d')
forecast_data = prophet.predict(future_data)
forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(5)
```

```python
pd.plotting.register_matplotlib_converters()
fig1 = prophet.plot(forecast_data)
```

```python
fig2 = prophet.plot_components(forecast_data)
```

```python
pro_clicks = df_clicks.rename(columns={'date':'ds', 'clicks':'y'})
pro_clicks
```

```python
from matplotlib import pyplot
```

```python
from fbprophet import Prophet
pd.plotting.register_matplotlib_converters()
model = Prophet()

model.fit(pro_clicks)
```

```python
# train set 20일 이후의 00일 날짜 생성
last_week = list()
for i in range(14, 21):
    last_week.append(['2021-04-%02d' % i])
last_week = pd.DataFrame(last_week, columns = ['ds'])
last_week['ds']= pd.to_datetime(last_week['ds'])
```

```python
last_week
```

```python
forecast = model.predict(last_week)
model.plot(forecast)
pyplot.show()
```

```python
# train set 마지막 1주 날짜 생성
last_week = list()
for i in range(21, 24):
    last_week.append(['2021-04-%02d' % i])
last_week = pd.DataFrame(last_week, columns = ['ds'])
last_week['ds']= pd.to_datetime(last_week['ds'])
```

```python
# 마지막 00일 제외시키기
train = pro_clicks.drop(pro_clicks.index[-5:])
y_true = pro_clicks['y'][-5:].values

# 모델 생성 후 학습
model = Prophet()
model.fit(train)

# train set 마지막 00일 날짜 생성
last_1year = list()
for i in range(16, 21):
    last_1year.append(['2021-04-%02d' % i])
last_1year = pd.DataFrame(last_1year, columns = ['ds'])
last_1year['ds']= pd.to_datetime(last_1year['ds'])

# 예측하고 비교하기
# 여기에서는 MAE를 살펴본다.
forecast = model.predict(last_1year)
y_pred = forecast['yhat'].values
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_true, y_pred)
print('MAE: %.3f' % mae)
```

```python
pyplot.plot(y_true, label='Expected')
pyplot.plot(y_pred, label='Predicted')
pyplot.title("Prophet - clicks")
pyplot.xlabel("days")
pyplot.xticks([0, 1, 2, 3, 4])
pyplot.ylabel("clicks")
pyplot.legend()
pyplot.show()
```
