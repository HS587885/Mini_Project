## DATA2 아로나민 골드 2021년 1월부터 2022년 2월까지 데이터

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data2.info()
```

```python
#data2['Date'] = pd.to_datetime(data2['Date'], format='%Y%m%d%')
data2['년월'] = data2['년'] + data2['월']
data2
```

```python
# 연도의 각 년도 별 탑승객 수 시각화

# 막대 그래프
plt.subplots(figsize=(24,12))
plt.rc('font', family='NanumBarunGothic') 
sns.barplot(data = data2, x = 'Date', y = '판매수량',  label="아로나민골드_판매금액")
```

```python
# 연도의 각 년도 별 탑승객 수 시각화

# 막대 그래프
plt.subplots(figsize=(15,6))
plt.rc('font', family='NanumBarunGothic') 
sns.barplot(data = data2, x = 'Date', y = '판매수량')
```

```python
plt.figure(figsize=(24, 12))
plt.plot(data2['Date'], data2['판매금액'], label="Sales amount")
plt.grid(True)
plt.legend(loc=5) # 범례 
plt.title("From 2021 TO 2022 AronaminGold Sales amount")
plt.xlabel("Timeline")
plt.ylabel("Sales amount") 
plt.show()
```

- 2021년 2월부터 4월까지 꾸준히 금액 증가
- 2021년 8월부터 9월까지 급격히 금액 증가


```python
plt.figure(figsize=(24, 12))
plt.rc('font', family='NanumBarunGothic') 
plt.plot(data2['Date'], data2['판매수량'], label="AronaminGold Sales quantity")
plt.grid(True)
plt.legend(loc=5) # 범례 
plt.title("From 2021 To 2022 the number of AronaminGold sold out")
plt.xlabel("Date")
plt.ylabel("Sales quantity") 
plt.show()
```

- 2021년 2월부터 3월 판매량 급격히 증가후 꾸준히 6월까지 상승

```python
data1['날짜'] = pd.to_datetime(data1['날짜'], format='%Y%m%d')
```

```python
plt.figure(figsize=(24, 12))
plt.plot(data1['날짜'], data1['전체'], label="AronaminGold")
plt.rc('font', family='NanumBarunGothic') 
plt.grid(True)
plt.legend(loc=5) # 범례 
plt.title("2021 to 2022")
plt.xlabel("Timeline")
plt.ylabel("Platform_totalCount") 
plt.show()
```

- 아로나민 골드에 대해 2월부터 4월까지 언급량이 많다고 볼 수 있다.

```python
plt.figure(figsize=(24, 12))
plt.plot(data1['날짜'], data1['트위터(RT 제외)'], label="Twitter without RT")
plt.plot(data1['날짜'], data1['블로그'], label="Blog")
plt.plot(data1['날짜'], data1['커뮤니티'], label="Community")
plt.plot(data1['날짜'], data1['인스타그램'], label="Instagram")
plt.plot(data1['날짜'], data1['뉴스'], label="News")




plt.grid(True)
plt.legend(loc=3) # 범례 
plt.title("AronaminGold by platforms")
plt.xlabel("TimeLine")
plt.ylabel("Platforms") 
plt.show()
```

-  아로나민 골드를 언급한 데이터에서 블로그가 가장 많고 그 다음이 커뮤티니라고 볼 수 있다

- 언급량이 많다고 해서 꼭 판매수량 및 팬매금액이 많아지거나 증가하는 건 아니다 하지만 언급량이 증가하고 판매 수량 및 금액이 함께 비례했을 경우에는 correlation을 가지게 하는 행사나 이벤트 또는 광고 켐페인이 있을 수 있다.
