```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# warning 없애기
import warnings
warnings.filterwarnings(action='ignore')
import platform
if platform.system() == 'Windows': # Windows
    plt.rc('font', family='Malgun Gothic')
```

# 데이터 불러오기

```python
data = pd.read_csv('광고집행 모니터링 201901-202203.csv', encoding = 'cp949')
data
```

# 횟수 데이터 분리

```python
data_count = data[data['항목']=='횟수']
data_count
```

# 일동제약의 매체별 광고 횟수

```python
#일동제약 라디오
ildong_radio = data_count[data_count['광고주']=='일동제약']['라디오'].sum()
#일동제약 신문
ildong_newspaper = data_count[data_count['광고주']=='일동제약']['신문'].sum()
#일동제약 TV 매체 Total
a = data_count[data_count['광고주']=='일동제약']['지상파'].sum()
b = data_count[data_count['광고주']=='일동제약']['케이블(닐슨추정보너스)'].sum()
c = data_count[data_count['광고주']=='일동제약']['종편(닐슨추정보너스)'].sum()
ildong_tv = a+b+c
#일동제약 잡지
ildong_magazine = data_count[data_count['광고주']=='일동제약']['잡지'].sum()
#일동제약 디지털동영상
ildong_digitalvideo = data_count[data_count['광고주']=='일동제약']['디지털동영상'].sum()
#일동제약 디지털배너
ildong_digitalbanner = data_count[data_count['광고주']=='일동제약']['디지털배너'].sum()
```

```python
print(f'라디오 : {ildong_radio} \n신문 : {ildong_newspaper} \nTV : {ildong_tv} \n잡지 : {ildong_magazine} \n디지털동영상 : {ildong_digitalvideo} \n디지털배너 : {ildong_digitalbanner}')
```

# 일동제약의 월별 광고횟수(19.01~22.02)

```python
data2 = data_count[data_count['광고주']=='일동제약']
data3 = data2.groupby(['연월']).sum().drop(['지상파','라디오','신문','잡지','케이블(닐슨추정보너스)','종편(닐슨추정보너스)','디지털동영상','디지털배너'], axis = 1)
data3 = data3.reset_index()
data3
```
